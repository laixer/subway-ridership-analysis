```python
import pandas
from ggplot import *
import scipy
import numpy as np
%matplotlib inline
```

```python
turnstile_data = pandas.read_csv('turnstile_weather_v2.csv')
turnstile_data['DATEn'] = pandas.to_datetime(turnstile_data['DATEn'])
turnstile_data.describe()
```

```python
turnstile_data.info()
```

```python
turnstile_data_by_date = turnstile_data[['DATEn', 'day_week', 'ENTRIESn_hourly']].groupby(['DATEn', 'day_week'], as_index=False).sum()
turnstile_data_by_date.describe()
```

```python
turnstile_data_by_date
```

```python
turnstile_data_by_date.info()
```

```python
p = ggplot(turnstile_data_by_date, aes(x='DATEn', y='ENTRIESn_hourly')) + \
    geom_point() + \
    geom_line() + \
    scale_x_date(breaks=date_breaks('7 days'), labels = date_format("%Y-%m-%d")) + \
    xlab('date') + \
    ylab('total entries') + \
    ggtitle('total entries by date')
p.draw()
```

Interesting to note that ridership on Mondays appears to be lower than on other weekdays. 

In this sample, ridership is lowest on Sundays, followed by Saturdays,

```python
turnstile_data_weather = turnstile_data[['DATEn', 'rain']].groupby(['DATEn'], as_index=False).aggregate(lambda x: sum(x) / float(len(x)))
#turnstile_data_weather
```

```python
ggplot(turnstile_data_weather, aes(x='DATEn', y='rain')) + \
    geom_point() + \
    geom_line() + \
    scale_x_date(breaks=date_breaks('7 days'), labels = date_format("%Y-%m-%d")) + \
    xlab('date')
```

```python
non_rainy_turnstile_data = turnstile_data[(turnstile_data['DATEn'] >= '2011-05-07') & (turnstile_data['DATEn'] <= '2011-05-13')]
non_rainy_turnstile_data
```

```python
rainy_turnstile_data = turnstile_data[(turnstile_data['DATEn'] >= '2011-05-14') & (turnstile_data['DATEn'] <= '2011-05-20')]
rainy_turnstile_data
```

```python
ggplot(non_rainy_turnstile_data[["ENTRIESn_hourly"]].reset_index(), aes(x='ENTRIESn_hourly')) \
    + geom_histogram(binwidth=200) \
    + xlab('hourly entries') \
    + ylab('count') \
    + ggtitle('histogram of entry counts for non-rainy period 5/7/2011 - 5/13/2011')
```

```python
ggplot(rainy_turnstile_data[["ENTRIESn_hourly"]].reset_index(), aes(x='ENTRIESn_hourly')) \
    + geom_histogram(binwidth=200) \
    + xlab('hourly entries') \
    + ylab('count') \
    + ggtitle('histogram of entry counts for rainy period 5/14/2011 - 5/20/2011')
```
