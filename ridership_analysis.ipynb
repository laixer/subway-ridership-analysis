```python
import pandas
from ggplot import *
%matplotlib inline
```

```python
turnstile_data = pandas.read_csv('turnstile_weather_v2.csv')
turnstile_data['DATEn'] = pandas.to_datetime(turnstile_data['DATEn'])
turnstile_data.describe()
```

```python
turnstile_data.info()
```

```python
turnstile_data_by_date = turnstile_data[['DATEn', 'ENTRIESn_hourly']].groupby('DATEn', as_index=False).sum()
turnstile_data_by_date.describe()
```

```python
turnstile_data_by_date.info()
```

```python
p = ggplot(turnstile_data_by_date, aes(x='DATEn', y='ENTRIESn_hourly')) + geom_point() + geom_line()# + stat_smooth()
p.draw()
```
