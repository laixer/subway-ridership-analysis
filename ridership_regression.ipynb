```python
import pandas
from ggplot import *
import scipy
import numpy as np
from sklearn import datasets, linear_model
from sklearn.cross_validation import train_test_split
%matplotlib inline
```

```python
turnstile_data = pandas.read_csv('turnstile_weather_v2.csv')
turnstile_data['DATEn'] = pandas.to_datetime(turnstile_data['DATEn'])
turnstile_data.describe()
```

```python
turnstile_data.info()
```

```python
stations = turnstile_data['station'].unique()
station_features = []
for station in stations:
    new_col = "station-" + station
    station_features.append(new_col)
    turnstile_data[new_col] = turnstile_data['station'] == station
turnstile_data.describe()
```

```python
conds = turnstile_data['conds'].unique()
cond_features = []
for cond in conds:
    new_col = "cond-" + cond
    cond_features.append(new_col)
    turnstile_data[new_col] = turnstile_data['conds'] == cond
turnstile_data.describe()
```

```python
msk = np.random.rand(len(turnstile_data)) < 0.8

turnstile_data_train = turnstile_data[msk]
turnstile_data_test = turnstile_data[~msk]
```

```python
features = ['weekday', 'day_week', 'hour'] + station_features + cond_features

regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(turnstile_data_train[features], turnstile_data_train['ENTRIESn_hourly'])

# The coefficients
print('Features:\n', features)
print('Coefficients: \n', regr.coef_)

for a, b in zip(features, regr.coef_):
    print '%-30s: %s' % (a, b)

# The mean square error
#print("Residual sum of squares: %.2f"
#      % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(turnstile_data_test[features], turnstile_data_test['ENTRIESn_hourly']))
```
